```python
import pandas as pd

path = './'
file_name = 'binf6111qc_patient.xlsx'
```

```python
patient_details = pd.read_excel(path + file_name, sheet_name = 0)
sample_details = pd.read_excel(path + file_name, sheet_name = 1)
sample_purity = pd.read_excel(path+file_name, sheet_name = 2)
patient_details.insert(0, 'id', range(1, 1 + len(patient_details)))
sample_details.insert(0, 'id', range(1, 1 + len(sample_details)))
sample_purity.insert(0, 'id', range(1, 1 + len(sample_purity)))
```
